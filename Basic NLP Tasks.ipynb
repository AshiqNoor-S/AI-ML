# Stemming

```python
porter = nltk.PorterStemmer()
stemmedwords = [porter.stem(t) for t in dist]
stemmedwords
```

# Lemmetization

```python
WNlemma = nltk.WordNetLemmatizer()
lemmetizedwords = [WNlemma.lemmatize(t) for t in dist]
lemmetizedwords
```

Here we can see unlike the stemmed words all the individual tokens are meaningful
